# DataAugmentation

Ce script à pour but de manipuler les sons COVID, en modifiant la hauteur du son, la rapidité ou encore le décalement de ce dernier, dans le but de multiplier nos sons COVID et entrainer au mieux notre modèle

```python
#Import de modules 

import numpy as np
import librosa
import IPython.display as ipd
import matplotlib.pyplot as plt
import soundfile as sf
import os
import random


from pydub import AudioSegment
```

## DataAugmentation un seul son

```python
# On importe le son

cov_file = "son_test/cov1.wav"
```

```python
# Lecture du son 
ipd.Audio(cov_file)
```

```python
# Enregistrement à l'aide de librosa

sound = librosa.core.load(cov_file)[0]


y, sr = librosa.load(cov_file)
```

```python
# Fonction d'affichage

def plot_time_series(data, titre):
    fig = plt.figure(figsize=(14, 8))
    plt.title(titre)
    plt.ylabel('Amplitude')
    plt.plot(np.linspace(0, 1, len(data)), data)
    plt.show()
```

```python
# Shifting Time

#Le but du shifting time est de décaler aléatoirement à droite ou à gauche le son, avec un nombre de secondes aléatoires
plot_time_series(sound, 'Basic sound')

data_roll = np.roll(sound, 10000)
plot_time_series(data_roll,'Sound roll')

# Ecoute du son Après Shifting time
ipd.Audio(data_roll, rate=sr)
```

```python
# Enregistrement 
sf.write('sound_shifted.wav', data_roll, sr, 'PCM_24')
```

```python
# Changing Pitch/Hauteur du son


plot_time_series(sound,'Basic Sound')

sound_pitch = librosa.effects.pitch_shift(y, sr, n_steps=4)

ipd.Audio(sound_pitch, rate=sr)


plot_time_series(sound_pitch,'Sound with changing pitch ')

# Ecoute du son Après Shifting time
ipd.Audio(sound_pitch, rate=sr)
```

```python
# Enregistrement 
sf.write('sound_pitched.wav', sound_pitch, sr, 'PCM_24')
```

```python
# Changing Speed/Rapidité

plot_time_series(sound,'Basic Sound')

sound_speed = librosa.effects.time_stretch(y, 0.5)
plot_time_series(sound_speed,'Sound Speed')


# Ecoute du son Après Shifting time
ipd.Audio(sound_speed, rate=sr)
```

```python
# Enregistrement 
sf.write('sound_speed.wav', sound_speed, sr, 'PCM_24')
```

## DataAugmentation l'ensemble du dataset COVID

```python
# Manipule chaque son en 3 sons différents, un son décalé, un son modifié au niveau de la hauteur et un son ralenti

path = os.listdir("audio-Cov-wav")

for file in path:
    sound = librosa.core.load("audio-Cov-wav/"+file)[0]
    y, sr = librosa.load("audio-Cov-wav/"+file)
    
    data_roll = np.roll(sound, 30000)
    sf.write("audio-augmented/"+file[:-4]+'shifted'+'.wav', data_roll, sr, 'PCM_24')
    
    sound_pitch = librosa.effects.pitch_shift(y, sr, n_steps=5)
    sf.write("audio-augmented/"+file[:-4]+'pitched''.wav', sound_pitch, sr, 'PCM_24')
    
    sound_speed = librosa.effects.time_stretch(y, 0.5)
    sf.write("audio-augmented/"+file[:-4]+'speed''.wav', sound_speed, sr, 'PCM_24')
    
```

```python
# Script final qui permets de multiplier par 5 sur chaque paramètres


path = os.listdir("audio-Cov-wav")

for file in path:
    sound = librosa.core.load("audio-Cov-wav/"+file)[0]
    y, sr = librosa.load("audio-Cov-wav/"+file)
    
    for i in range(5):
        num = random.randint(10000,80000)
        data_roll = np.roll(sound, num)
        sf.write("audio-augmented/"+file[:-4]+'-shifted-'+str(i)+'.wav', data_roll, sr, 'PCM_24')
    for i in range(5):
        num2 = random.randint(1,50)
        sound_pitch = librosa.effects.pitch_shift(y, sr, n_steps=num2)
        sf.write("audio-augmented/"+file[:-4]+'-pitched-'+str(i)+'.wav', sound_pitch, sr, 'PCM_24')    
    for i in range(5):
        num3 = random.uniform(0.1,2.3)
        sound_speed = librosa.effects.time_stretch(y, num3)
        sf.write("audio-augmented/"+file[:-4]+'-speed-'+str(i)+'.wav', sound_speed, sr, 'PCM_24')
    
```
